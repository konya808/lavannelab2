```python
import numpy as np
import matplotlib.pyplot as plt

resistor_value = 2000
capacitor_value = 2e-6
voltage_input = 5

tau_calc = resistor_value * capacitor_value
print("The time constant (tau) is:", tau_calc, "seconds")

time_array = np.linspace(0, 0.02, 1000)

capacitor_voltage_discharge = voltage_input * np.exp(-time_array / (resistor_value * capacitor_value))

plt.plot(time_array, capacitor_voltage_discharge)
plt.xlabel("Time in seconds")
plt.ylabel("Voltage in Volts")
plt.title("My RC Discharging Curve Plot!!!")
plt.grid(True)
plt.show()
```
