```python
import numpy as np
import matplotlib.pyplot as plt

R = 2000      # Ohms (Changed from 1000)
C = 2e-6      # Farads (Changed from 1e-6)
V = 5         # Volts

# Calculate the time constant
tau = R * C
print(f"Time constant (tau) = {tau:.3e} seconds")

# Adjust time range to better show discharge based on new tau
t = np.linspace(0, 5 * tau, 1000) # Plot over 5 time constants for better visualization

# Discharging capacitor voltage formula: Vc(t) = V * e^(-t / (R*C))
Vc = V * np.exp(-t / (R * C))

plt.plot(t, Vc)
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.title("RC Discharging Curve")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Using the same R, C, V values from the previous cell
R_charge = R # Ohms
C_charge = C # Farads
V_charge = V # Volts

# Calculate the time constant for charging
tau_charge = R_charge * C_charge
print(f"Time constant for charging (tau) = {tau_charge:.3e} seconds")

# Adjust time range for charging plot (typically 5 time constants)
t_charge = np.linspace(0, 5 * tau_charge, 1000)

# Charging capacitor voltage formula: Vc(t) = V * (1 - e^(-t / (R*C)))
Vc_charge = V_charge * (1 - np.exp(-t_charge / (R_charge * C_charge)))

plt.plot(t_charge, Vc_charge)
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.title("RC Charging Curve")
plt.grid(True)
plt.show()
```
